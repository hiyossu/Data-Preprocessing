```python
import pandas as pd
import numpy as np
import io

df = pd.read_csv('/content/Messy_Ecommerce_Orders.csv')
df
```

```python
#Standardization of order_date
df['order_date'] = pd.to_datetime(df['order_date'], format='mixed', dayfirst = False, errors='coerce')

df
```

```python
#Standardization of ship_date
df['ship_date']  = pd.to_datetime(df['ship_date'], format='mixed', dayfirst = False, errors='coerce')
df
```

```python
#Finding all possible unique instances of the name of countries

print(df['country'].unique())
```

```python
# Normalize country

df['country'] = df['country'].astype(str).str.strip()
UNITED_STATES = ["U.S.", "United States", "US"]
PHILIPPINES = ["P.H.", "Philippines", "PH"]
MALAYSIA = ["M.A.", "Malaysia", "MY"]
SINGAPORE = ["S.G.", "Singapore", "SG"]

df['country'] = df['country'].replace(UNITED_STATES, 'United States')
df['country'] = df['country'].replace(PHILIPPINES, 'Philippines')
df['country'] = df['country'].replace(MALAYSIA, 'Malaysia')
df['country'] = df['country'].replace(SINGAPORE, 'Singapore')

df
```

```python
print(df['payment_method'].unique())
```

```python
#Normlize payment_method column

df['payment_method'] = df['payment_method'].astype(str).str.strip()

PAYPAL = ['PayPal', 'paypal ']
CREDIT_CARD = [ 'credit card' , 'CC', 'CreditCard']
BANK_TRANSFER = ["gcash", "G-Cash", ]
CASH = ["Cash", "cash", "CASH"]

df['payment_method'] = df['payment_method'].replace(PAYPAL, 'PayPal')
df['payment_method'] = df['payment_method'].replace(CREDIT_CARD, 'Credit Card')
df['payment_method'] = df['payment_method'].replace(BANK_TRANSFER, 'Bank Transfer')
df['payment_method'] = df['payment_method'].replace(CASH, 'Cash')

df

```

```python
df['item_category'] = df['item_category'].str.upper()
df
```

```python
df['quantity'] = df['quantity'].abs()
df['unit_price'] = df['unit_price'].abs()

df['total_amount'] = df['total_amount'].str.replace('$', '', regex=False)
df['total_amount'] = df['total_amount'].str.replace('.', '', regex=False)
df['total_amount'] = df['total_amount'].str.replace(',', '', regex=False)
df['total_amount'] = df['total_amount'].str.replace('N/A', '0', regex=False)

```

```python
df['total_amount'] = df['total_amount'].astype(float).abs()
df
```

```python
df['order_status'] = df['order_status'].str.upper()

df
```

```python
# standardize returns as strings
df['is_returned'] = df['is_returned'].astype(str).str.strip()

NO = ['no', 'NO', 'No', '0', 'FALSE', 'false', 'N', 'n']
YES = ['yes', 'YES', 'Yes', '1', 'TRUE', 'true', 'Y', 'y']

df['is_returned'] = df['is_returned'].replace(NO, 'FALSE')
df['is_returned'] = df['is_returned'].replace(YES, 'TRUE')

df
```

```python
# checks if df['order_date'] and df['shipdate'] are invalid
date_comparison = pd.to_datetime(df['order_date']).dt.date > pd.to_datetime(df['ship_date']).dt.date

if date_comparison.any():
  problematic_orders = df[date_comparison]
  print('Error: The following orders have an order_date after the ship_date:')
  print(problematic_orders[['order_id', 'order_date', 'ship_date']])
else:
  print('No orders found where order_date is after ship_date.')
```

```python
# checks if any order_ids are duplicated
order_id_comparison = df['order_id'].duplicated(keep=False)

if order_id_comparison.any():
  problematic_orders = df[order_id_comparison]
  print('Error: The following orders have the same order_id:')
  print(problematic_orders[['order_id']])
else:
  print('No orders found where order_id is the same.')
```

```python
# IQR calculator

Q1 = df['total_amount'].quantile(0.25)
Q3 = df['total_amount'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f"Q1 (First Quartile): {Q1}")
print(f"Q3 (Third Quartile): {Q3}")
print(f"IQR (Interquartile Range): {IQR}")
print(f"Lower Bound for Outliers: {lower_bound}")
print(f"Upper Bound for Outliers: {upper_bound}")
```
